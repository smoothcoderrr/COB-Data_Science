```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("D:\data sets\dataset - netflix1.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.dtypes
```

```python
data['date_added']=pd.to_datetime(data['date_added'],format='mixed')
data['date_added'].head()
```

```python
data.isnull().sum()
```

```python
data=data[((data['date_added'].dt.year)>=data['release_year'])]
```

```python
data['title']=data['title'].drop_duplicates()
```

```python
data[data['title'].duplicated()]
```

```python
data.drop_duplicates(['title'],inplace=True)
```

```python
sns.countplot(x='type',data=data,palette='Set2')
plt.title("Count VS Type")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country', order=data['country'].value_counts().index[0:10],data=data,palette='Set1')
plt.title("Country Wise Content On NetFlix")
```

```python
movie_countries=data[data['type']=='Movie']
tv_show_countries=data[data['type']=='TV Show']
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=data['country'].value_counts().index[0:10],data=movie_countries,palette='Set3')
plt.title('Top 10 Countries Producing Movies On Netflix')

plt.figure(figsize=(12,6))
sns.countplot(y='country',order=data['country'].value_counts().index[0:10],data=tv_show_countries,palette='Set3')
plt.title('Top 10 Countries Producing TV_Show On Netflix')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='rating', order=data['rating'].value_counts().index[0:10], data= data,palette='Set2')
plt.title("Ratings of show Vs")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='release_year',order=data['release_year'].value_counts().index[0:20],data=data,palette='Set1')
plt.title("Content Released Year On NetFlix VS Count")
```

```python
counts_data = data.groupby(['release_year', 'type']).size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.barplot(x='release_year', y='count', hue='type' ,order=data['release_year'].value_counts().index[0:20],data=counts_data, palette='Set2')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Count of Movie Types by Release Year')
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
